```python
from PIL import ImageFont
```

```python
im = Image.open('src/Images/frame.jpg')
```

```python
draw = ImageDraw.Draw(im)
```

```python
step = 30
```

```python
for i in range(0, FRAME_HEIGHT, step):
    draw.line(xy=(0, i, FRAME_WIDTH, i), fill='green', width=2)
    
for i in range(0, FRAME_WIDTH, step):
    draw.line(xy=(i, 0, i, FRAME_HEIGHT), fill='green', width=2)
```

```python
obs = Image.open('src/Images/obstacles.png')
```

```python
obs = (~np.array(obs).sum(axis=2).astype(bool)).astype(int)
```

```python
for i in range(int(FRAME_HEIGHT / step)):
    for j in range(int(FRAME_WIDTH / step)):
        x1, x2 = i * step, (i + 1) * step
        y1, y2 = j * step, (j + 1) * step
        if obs[x1 : x2, y1 : y2].sum() > step * step / 2:
            draw.rectangle((y1+1, x1+1, y2, x2), fill='grey')
```

```python
font = ImageFont.load_default()
```

```python
for i in range(int(FRAME_HEIGHT / step)):
    for j in range(int(FRAME_WIDTH / step)):
        if i % 10 == 0 and j % 10 == 0:
            x1, y1 = i * step, j * step
            draw.text(xy=(y1+1, x1+1), text='{i}, {j}'.format(i=i, j=j), fill='white', font=font)
```
